## IMPORT NECESSARY LIBRARIES :

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## IMPORT THE DATASET :

```python
sales = pd.read_csv("DVA on sales data\SalesData.csv")
```

```python
sales.head()
```

#### 1. Compare Sales by region for 2016 with 2015 using bar chart.

```python
region_sales = sales.groupby("Region")["Sales2015", "Sales2016"].sum()
```

```python
region_sales
```

```python
region_sales.plot(kind = "bar", figsize = (8, 8))
plt.ylabel("Sales")
plt.title("Sales by Region")
plt.show()
```

#### We can conclude that sales in 2016 is more as compared to 2015 in all regions. East region has contributedd the maximum.

#### 2. What are the contributing factors to the sales for each region in 2016. Visualize it using a Pie Chart.

```python
sales_pie = sales.groupby("Region")["Sales2016"].sum()
```

```python
sales_pie
```

```python
plt.pie(sales_pie, autopct = "%1.0f%%", labels = ["Central", "East", "West"], shadow = True, explode = [0.0, 0.1, 0.0], colors = ["r", "g", "b"])
plt.title("Contributing Factors to the Sales for each Region in 2016")
plt.show()
```

#### East region has contributed the maximum.

#### 3. Compare the total sales of 2015 and 2016 with respect to Region and Tiers.

```python
sales_region_tier = sales.groupby(["Region", "Tier"])["Sales2015", "Sales2016"].sum()
sales_region_tier
```

```python
sales_region_tier.plot(kind = "bar", figsize = (8, 8))
plt.ylabel("Sales")
plt.title("Total Sales of 2015 and 2016 with respect to Region and Tiers")
plt.show()
```

#### We can conclude that east region and high tier in 2016 has contributed the maximum. Also sales in 2015 and 2016 both is maximum in East region and High tier.

#### 6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as-
####  Jan - Mar : Q1
####  Apr - Jun : Q2
####  Jul - Sep : Q3
####  Oct - Dec : Q4

```python
sales.head()
```

```python
month = sales["Month"]
```

```python
quarter = []
for x in month:
    if x in ["Jan", "Feb", "Mar"]:
        quarter.append("Q1")
    elif x in ["Apr", "May", "Jun"]:
        quarter.append("Q2")
    elif x in ["Jul", "Aug", "Sep"]:
        quarter.append("Q3")
    else:
        quarter.append("Q4")
```

```python
sales["Qtr"] = pd.Series(quarter)
```

```python
sales.head()
```

#### 7. Compare Qtr wise sales in 2015 and 2016 in a bar plot.

```python
sales_qtr = sales.groupby(["Qtr"])["Sales2015", "Sales2016"].sum()
```

```python
sales_qtr
```

```python
sales_qtr.plot(kind = "bar", figsize = (8,6))
plt.ylabel("Sales")
plt.title("Qtr wise sales in 2015 and 2016")
plt.show()
```

#### 8. Determine the composition of Qtr wise sales in and 2016 with regards to all the Tiers in a pie chart.
#### (Draw 4 pie charts representing a Quarter for each Tier)

```python
qtr_pivot = sales.pivot_table(index = "Qtr", columns = "Tier", values = "Sales2016")
```

```python
qtr_pivot
```
